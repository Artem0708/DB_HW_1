```python
import pandas as pd
```

```python
EXCEL_FILE = 'customer_and_transaction.xlsx'
```

Загружаем и смотрим данные

```python
transactions_df = pd.read_excel(EXCEL_FILE, sheet_name='transaction')
customers_df = pd.read_excel(EXCEL_FILE, sheet_name='customer')
```

```python
transactions_df.info()
```

```python
customers_df.info()
```

Группируем столбцы по таблицам

### transactions

```python
transactions_final = transactions_df[[
    'transaction_id', 'product_id', 'customer_id', 
    'transaction_date', 'online_order', 'order_status', 'list_price'
]].copy()
```

```python
transactions_final.info()
```

### products

```python
products_final = transactions_df[[
    'product_id', 'brand', 'product_line', 'product_class', 'product_size', 'standard_cost'
]].drop_duplicates('product_id').copy()
```

```python
products_final.info()
```

### customers

```python
customers_final = customers_df[[
    'customer_id', 'first_name', 'last_name', 'gender', 'DOB',
    'job_industry_category', 'job_title', 'wealth_segment',
    'deceased_indicator', 'owns_car', 'address'
]].copy()
```

```python
customers_final.info()
```

### customer_addresses

```python
customer_addresses_final = customers_df[[
    'country', 'state', 'postcode', 'address', 'property_valuation'
]].copy()
```

```python
customer_addresses_final = customer_addresses_final.drop_duplicates().reset_index(drop=True)
```

```python
customer_addresses_final['address_id'] = customer_addresses_final.index + 1
```

```python
customer_addresses_final = customer_addresses_final[['address_id', 'country', 'state', 'postcode', 'address', 'property_valuation']]
```

```python
customer_addresses_final.info()
```

### master

```python
master_final = transactions_df[['transaction_id', 'product_id', 'customer_id']].copy()
```

```python
master_final.info()
```

```python
master_final = master_final.merge(
    customers_df[['customer_id', 'address']], 
    on='customer_id', 
    how='left'
)
```

```python
master_final.info()
```

```python
unique_addresses = customer_addresses_final[['address', 'address_id']].drop_duplicates(subset=['address'])

master_final = master_final.merge(
    unique_addresses, 
    on='address', 
    how='left'
)
```

```python
master_final.info()
```

```python
# Оставляем только нужные колонки
master_final = master_final[['transaction_id', 'product_id', 'customer_id', 'address_id']]
```

```python
master_final['row_id'] = master_final.index + 1
master_final = master_final[['row_id', 'transaction_id', 'product_id', 'customer_id', 'address_id']]
```

```python
master_final.info()
```

В исходных данных на вкладке customers не существует клиента с ID = 5034, в то время как в транзакциях есть 3 строки с этим ID.
При загрузке DBeaver выдает ошибку.

В силу несущественности объема этих данных в общем объеме (3 наблюдения из 20000), удалим данные.

```python
master_clean = master_final[master_final['customer_id'] != 5034].copy()
```

Сохраняем в формате Excel для дальнейшей загрузки через DBeaver

```python
transactions_final.to_csv('transactions.csv', index=False, encoding='utf-8')
products_final.to_csv('products.csv', index=False, encoding='utf-8')
customers_final.to_csv('customers.csv', index=False, encoding='utf-8')
customer_addresses_final.to_csv('customer_addresses.csv', index=False, encoding='utf-8')
master_clean.to_csv('master.csv', index=False, encoding='utf-8')
```
